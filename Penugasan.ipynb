```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, filters
from skimage.filters.rank import gradient
from skimage.morphology import disk

# ===================== 1⃣ AMBIL GAMBAR BERWARNA =====================
image_color = io.imread("a.png")  # Mengambil gambar dari perangkat

# ===================== 2⃣ TERAPKAN BERBAGAI OPERATOR DETEKSI TEPI =====================
def apply_edge_detection(image, operator):
    """Menerapkan operator deteksi tepi pada setiap kanal warna (R, G, B)."""
    red = operator(image[:, :, 0])
    green = operator(image[:, :, 1])
    blue = operator(image[:, :, 2])
    return np.stack((red, green, blue), axis=2)

sobel_color = apply_edge_detection(image_color, filters.sobel)
roberts_color = apply_edge_detection(image_color, filters.roberts)
prewitt_color = apply_edge_detection(image_color, filters.prewitt)
kirsch_color = apply_edge_detection(image_color, lambda img: gradient(img, disk(1)))

# ===================== 3⃣ VISUALISASI HASIL (FORMAT 2 KOLOM, 5 BARIS) =====================
fig, axes = plt.subplots(5, 2, figsize=(10, 20))
ax = axes.ravel()

# ----- Baris 1: Gambar Asli -----
ax[0].imshow(image_color)
ax[0].set_title("Gambar Asli")
ax[0].axis("off")
ax[1].axis("off")

# ----- Baris 2: Sobel -----
ax[2].imshow(sobel_color)
ax[2].set_title("Sobel")
ax[2].axis("off")
ax[3].axis("off")

# ----- Baris 3: Roberts -----
ax[4].imshow(roberts_color)
ax[4].set_title("Roberts")
ax[4].axis("off")
ax[5].axis("off")

# ----- Baris 4: Prewitt -----
ax[6].imshow(prewitt_color)
ax[6].set_title("Prewitt")
ax[6].axis("off")
ax[7].axis("off")

# ----- Baris 5: Kirsch -----
ax[8].imshow(kirsch_color)
ax[8].set_title("Kirsch")
ax[8].axis("off")
ax[9].axis("off")

plt.tight_layout()
plt.show()
```
